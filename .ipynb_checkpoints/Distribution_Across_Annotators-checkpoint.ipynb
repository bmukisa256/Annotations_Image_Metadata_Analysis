```python
import pandas as pd
import os
```

```python
# Load CSV files into separate DataFrames
sample_annotations_df = pd.read_csv('sample_annotations_df.csv')
chris_annotations_df = pd.read_csv('chris_annotations_df.csv')
judaya_annotations_df = pd.read_csv('judaya_annotations_df.csv')
kati_annotations_df = pd.read_csv('kati_annotations_df.csv')
okwa_annotations_df = pd.read_csv('okwa_annotations_df.csv')
sally_annotations_df = pd.read_csv('sally_annotations_df.csv')

# New column for the annotator's name
sample_annotations_df['Annotator'] = 'Sample'
chris_annotations_df['Annotator'] = 'Chris'
judaya_annotations_df['Annotator'] = 'Judaya'
kati_annotations_df['Annotator'] = 'Kati'
okwa_annotations_df['Annotator'] = 'Okwa'
sally_annotations_df['Annotator'] = 'Sally'

# Concatenate
merged_df = pd.concat([sample_annotations_df, chris_annotations_df, judaya_annotations_df, kati_annotations_df, okwa_annotations_df,sally_annotations_df], ignore_index=True)
```

```python
annotator_counts = merged_df['Annotator'].value_counts()
print(annotator_counts)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

annotator_counts = merged_df['Annotator'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=annotator_counts.index, y=annotator_counts.values, alpha=0.8)
plt.title('Annotator Counts')
plt.xlabel('Annotator')
plt.ylabel('Count')
plt.show()
```

```python
# Unique category names for each annotator
sample_categories = sample_annotations_df['category_name'].unique()
chris_categories = chris_annotations_df['category_name'].unique()
judaya_categories = judaya_annotations_df['category_name'].unique()
kati_categories = kati_annotations_df['category_name'].unique()
okwa_categories = okwa_annotations_df['category_name'].unique()
sally_categories = sally_annotations_df['category_name'].unique()

# Print the results
print("Sample categories:", sample_categories)
print("Chris categories:", chris_categories)
print("Judaya categories:", judaya_categories)
print("Kati categories:", kati_categories)
print("Okwa categories:", okwa_categories)
print("Sally categories:", sally_categories)
```

```python
# Get the count of 'Trophozoite' annotations for each annotator
sample_trophozoite_count = sample_annotations_df[sample_annotations_df['category_name'] == 'Trophozoite'].shape[0]
chris_trophozoite_count = chris_annotations_df[chris_annotations_df['category_name'] == 'Trophozoite'].shape[0]
judaya_trophozoite_count = judaya_annotations_df[judaya_annotations_df['category_name'] == 'Trophozoite'].shape[0]
kati_trophozoite_count = kati_annotations_df[kati_annotations_df['category_name'] == 'Trophozoite'].shape[0]
okwa_trophozoite_count = okwa_annotations_df[okwa_annotations_df['category_name'] == 'Trophozoite'].shape[0]
sally_trophozoite_count = sally_annotations_df[sally_annotations_df['category_name'] == 'Trophozoite'].shape[0]

# Print the results
print("Sample Trophozoite count:", sample_trophozoite_count)
print("Chris Trophozoite count:", chris_trophozoite_count)
print("Judaya Trophozoite count:", judaya_trophozoite_count)
print("Kati Trophozoite count:", kati_trophozoite_count)
print("Okwa Trophozoite count:", okwa_trophozoite_count)
print("Sally Trophozoite count:", sally_trophozoite_count)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame with the counts
trophozoite_counts = pd.DataFrame({
    'Annotator': ['Sample', 'Chris', 'Judaya', 'Kati', 'Okwa', 'Sally'],
    'Count': [sample_trophozoite_count, chris_trophozoite_count, judaya_trophozoite_count, kati_trophozoite_count, okwa_trophozoite_count, sally_trophozoite_count]
})

# Create a bar plot with Seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x='Annotator', y='Count', data=trophozoite_counts)
plt.title('Distribution of Trophozoite Annotations')
plt.xlabel('Annotator')
plt.ylabel('Count')
plt.show()
```

```python
object_cols = merged_df.select_dtypes(['object']).columns
print(object_cols)
```

```python
# Remove file_name column
merged_df = merged_df.drop('file_name', axis=1)
merged_df = merged_df.drop('image_id', axis=1)
merged_df = merged_df.drop('category_id', axis=1)
#merged_df = merged_df.drop('Annotator', axis=1)

# Categorical columns to numeric representation
merged_df['category_name'] = pd.Categorical(merged_df['category_name']).codes
merged_df['Annotator'] = pd.Categorical(merged_df['Annotator']).codes

# Binary labels to numeric values
merged_df['is_thick'] = pd.to_numeric(merged_df['is_thick'], errors='coerce')
merged_df['is_thin'] = pd.to_numeric(merged_df['is_thin'], errors='coerce')
```

```python
import pandas as pd

# Calculate correlation between annotation columns
correlation_matrix = merged_df.corr()
correlation_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation matrix
corr_matrix = merged_df.corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```
